```python
class StudentList:

    def __init__(self):
        self.studentlist = []

    def add_details(self,reg_no,name,phone,email,marks1,marks2,marks3,marks4,marks5):

        details = {
            'Reg No' : reg_no,
            'Name' : name,
            'Phone' : phone,
            'Email' : email,
            'Marks' : {'Python': marks1 ,'Excel': marks2,'RM' : marks3,'Statistics': marks4,'Mathematics': marks5}
        }
        self.studentlist.append(details)

    def disp_details(self):
        for i in self.studentlist:
            print(i)
            print()
            print('Register No::',i['Reg No'])
            print('Name:',i['Name'])
            print('Phone:',i['Phone'])
            print('Email:',i['Email'])
            print('Python Marks:',i['Marks']['Python'])
            print('Excel Marks:',i['Marks']['Excel'])
            print('RM Marks:',i['Marks']['Excel'])
            print('Statistics Marks:',i['Marks']['Statistics'])
            print('Mathematics Marks:',i['Marks']['Statistics'])
            print()
        print(self.studentlist)
        print()

    def edit_details(self,reg_no):
        self.found = False
        for i in self.studentlist:
            if i['Reg No'] == reg_no:
                self.found = True
                print('Editing details of student with register no:',reg_no)
                self.new_reg = input('Enter the new register no.(Press enter if no change)').strip().title()
                self.new_name = input('Enter the new name(Press enter if no change)').strip().title()
                self.new_phone = input('Enter the new phone no.(Press enter if no change)').strip()
                self.new_email = input('Enter the new email id.(Press enter if no change)').strip().lower()
                self.new_marks1 = input('Enter the new marks for Python (Press enter if no change)').strip()
                self.new_marks2 = input('Enter the new marks for Excel (Press enter if no change)').strip()
                self.new_marks3 = input('Enter the new marks for RM (Press enter if no change)').strip()
                self.new_marks4 = input('Enter the new marks for Statistics (Press enter if no change)').strip()
                self.new_marks5 = input('Enter the new marks for Mathematics (Press enter if no change)').strip()

                if self.new_name:
                    i['Name'] = self.new_name
                if self.new_phone:
                    i['Phone'] = self.new_phone
                if self.new_email:
                    i['Email'] = self.new_email
                if self.new_marks1:
                    i['Marks']['Python'] = self.new_marks1
                if self.new_marks2:
                    i['Marks']['Excel'] = self.new_marks2
                if self.new_marks3:
                    i['Marks']['RM'] = self.new_marks3
                if self.new_marks4:
                    i['Marks']['Statistics'] = self.new_marks4
                if self.new_marks5:
                    i['Marks']['Mathematics'] = self.new_marks5
                print('Student Information updated successfully!!')
        if not self.found:
            print('Register No. not found!!')

    def delete_details(self,reg_no):
        reg_no = reg_no.strip().title()
        self.found = False
        for i in self.studentlist:
            if i['Reg No'] == reg_no:
                self.found = True
                self.studentlist.remove(i)
                print('Register No. deleted successfully!!')

        if not self.found:
            print('Register No. not found!!')

    def search_details(self,reg_no):
        reg_no = reg_no.strip().title()
        self.found = False
        self.index = 0
        for i in self.studentlist:
            if i['Reg No'] == reg_no:
                self.found = True
                break
            self.index += 1
        if self.found == True:    
            print('Register No. is in the list.')
            print('The details for entered register no. is')
            print(self.studentlist[self.index]['Reg No'])
            print(self.studentlist[self.index]['Name'])
            print(self.studentlist[self.index]['Phone'])
            print(self.studentlist[self.index]['Email'])
        else:
            print('Register No. NOT found!')

student = StudentList()

while True:

    print('MENU FOR 1ST YEAR MSC DATA SCIENCE CLASS LIST')
    print()
    print('1. Add details')
    print('2. Display Details')
    print('3. Edit existing details')
    print('4. Delete a contact')
    print('5. Search for a contact')
    print('6. Exit')
    print()

    choice = input('Enter your choice')
    print('Your entered choice is:',choice)

    if int(choice)== 1:
        reg_no = input('Enter the register no.').strip().title()
        name = input('Enter the name:').strip().title()
        phone = input('Enter the phone no.:').strip()
        email = input('Enter the email id:').strip().lower()
        marks1 = input('Enter marks for Python:').strip()
        marks2 = input('Enter marks for Excel:').strip()
        marks3 = input('Enter marks for RM:').strip()
        marks4 = input('Enter marks for Statistics:').strip()
        marks5 = input('Enter marks for Mathematics:').strip()
        print()
        student.add_details(reg_no,name,phone,email,marks1,marks2,marks3,marks4,marks5)
    elif int(choice)== 2:
        student.disp_details()
    elif int(choice)== 3:
        reg_no = input('Enter the register no. to be edited:').strip().title()
        student.edit_details(reg_no)
    elif int(choice) == 4:
        reg_no = input('Enter the register no. to be deleted:').strip().title()
        student.delete_details(reg_no)
    elif int(choice) == 5:
        reg_no = input('Enter the register no. to be searched:').strip().title()
        student.search_details(reg_no)
    elif int(choice) == 6:
        break
    else:
        print('Invalid selection!!')
```
