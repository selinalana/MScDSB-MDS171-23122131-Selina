```python
class stack(): #Class definition

    def __init__(self): #init function to define a list to store the stack
        self.stack = []

    def push_item(self,item): #function to push an item to the stack
        self.stack.append(item)
        print("Item",item,"pushed successfully to the stack!!")
        print()

    def pop_item(self): #function to pop the topmost element from the stack as it is LIFO
        self.stack.pop()
        print("Item successfully popped from the stack!!")
        print()

    def disp_stack(self): #function to display the stack
        print("The stack items are:",self.stack)
        print()

# Get the current size of the stack
    def getSize(self):
        print("The size of the stack is:",len(self.stack))
        print()

# Get the top item of the stack
    def peek(self):
        if not self.isEmpty():
            return self.stack[-1]
        else:
            raise Exception("Peeking from an empty stack")

# Check if the stack is empty
    def isEmpty(self):
        return len(self.stack) == 0
    
stack_obj = stack()

while True:
    print("STACK OPERATIONS")
    print("1. Push an item")
    print("2. Pop an item")
    print("3. Print Stack items")
    print("4. Print Stack Size")
    print("5. Print Top Stack Item")
    print("6. Check if stack is empty")
    print("7. Exit")
    print()

    choice = int(input("Enter your choice").strip())

    if choice ==1:
        print("Your entered choice is:",choice)
        num = input("Enter an item to push")
        stack_obj.push_item(num)
    elif choice ==2:
        print("Your entered choice is:",choice)
        stack_obj.pop_item()
    elif choice ==3:
        print("Your entered choice is:",choice)
        stack_obj.disp_stack()
    elif choice ==4:
        print("Your entered choice is:",choice)
        stack_obj.getSize()
    elif choice ==5:
        print("Your entered choice is:",choice)
        print("Top item in the stack is:",stack_obj.peek())
        print()
    elif choice ==6:
        print("Your entered choice is:",choice)
        if stack_obj.isEmpty():
            print("Stack is empty.")
        else:
            print("Stack is not empty.")
        print()
    elif choice ==7:
        break
    else:
        print("Invalid choice!")
```
