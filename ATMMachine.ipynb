```python
def main():
    # Lists to store account data
    pinCode = ["1234", "1999", "2424", "1985", "5555"]
    accountHoldersName = ["ansel", "george", "alwin", "selina lana", "lingesh"]
    accountNumber = ['1353', '199281', "182838", "185597", "667432"]
   
    # Dictionary to store account balances (using account numbers as keys)
    balance = {'1353': 567000, '199281': 21873, '182838': 2341871, '185597': 275638, '667432': 91820}

    max_attempts = 3

    while True:
        print("""
        === Welcome To Simple ATM System ===""")
       
        # Input for the user's name (converted to lowercase for case-insensitive matching)
        inputName = input("Enter Your Name: ").lower()

        if inputName in accountHoldersName:
            # Find the index of the account holder's name in the list
            index = accountHoldersName.index(inputName)
           
            inputPin = input("Enter Pin Number: ")
           
            # Check if the input PIN matches the stored PIN for the account
            if inputPin == pinCode[index]:
                while True:  # Allow multiple operations without exiting
                    print("\nYour account number is:", accountNumber[index])
                    print("Your account balance is: Rs.", balance[accountNumber[index]])
                   
                    drawOrDeposit = input("\nDo you want to draw or deposit cash (draw/deposit/no): ")
                   
                    if drawOrDeposit == "draw":
                        amount = int(input("Enter the amount you want to draw: "))
                       
                        # Check if the account has sufficient balance for the withdrawal
                        if amount > balance[accountNumber[index]]:
                            print("Insufficient balance.")
                        else:
                            # Update the balance after the withdrawal
                            balance[accountNumber[index]] -= amount
                            print("Your available balance is:", balance[accountNumber[index]])
                   
                    elif drawOrDeposit == "deposit":
                        amount = int(input("Enter the amount you want to deposit: "))
                       
                        # Update the balance after the deposit
                        balance[accountNumber[index]] += amount
                        print("Your available balance is:", balance[accountNumber[index]])
                   
                    elif drawOrDeposit == "no":
                        print("\nThank you for using this Simple ATM System.")
                        break  # Exit the operation loop
                   
                    else:
                        print("Invalid choice. Please enter 'draw', 'deposit', or 'no'.")
               
                # Exit the login loop
                break
           
            else:
                print("Invalid PIN.")
       
        else:
            print("Account not found.")
       
        max_attempts -= 1
        if max_attempts == 0:
            print("Maximum login attempts reached. Exiting.")
            break

main()
```
