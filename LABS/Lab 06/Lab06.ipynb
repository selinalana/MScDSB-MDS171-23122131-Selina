```python
filename = "23122131.csv"

def process(filename):
    item_info = {} #defining an empty dictionary

    with open(filename, "r+") as file: #opening the file in read mode
        for line in file:
            parts = line.strip().split(",") #removing whitespaces and splitting at ,
            item, quantity, price = parts[0], int(parts[1]), int(parts[2]) #type-casting and segregating

            if item not in item_info: #checking the presence of an item in the dictionary
                item_info[item] = { 'Total Quantity': quantity,
                                    'Minimum Price': price,
                                    'Maximum Price': price,
                                    'Total Cost':  quantity* price
                                   }
            else:
                item_info[item] ['Total Quantity'] += quantity
                item_info[item] ['Total Cost'] += quantity*price
                if price < item_info[item]["Minimum Price"]:
                    item_info[item]["Minimum Price"] = price
                if price > item_info[item]["Maximum Price"]:
                    item_info[item]["Maximum Price"] = price
    
    return item_info
```

```python
def analysis(item_info):
    print("Item Name | Total Quantity | Minimum Price | Total Cost \n")
    for item, data in item_info.items():
        print(f"{item} | {data['Total Quantity' ]} | {data['Minimum Price']} | {data['Maximum Price']} | {data['Total Cost']}")
```

```python
# function call
item_info = process(filename)
analysis(item_info)
```
